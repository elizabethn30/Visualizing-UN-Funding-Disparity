```python
allocation_df = pd.read_csv("/Workspace/Users/elizabethnguyen@ufl.edu/cbpf_allocation_by_org_type.csv")
```

```python
#Turn the columns into floats to prepare for operations
total_needs_for_all_cluster["In Need"] = total_needs_for_all_cluster["In Need"].astype(float)
allocation_df["Total Allocation"] = allocation_df["Total Allocation"].astype(float)

#Group the two dataframes by country
national_needs = total_needs_for_all_cluster.groupby("Country ISO3", as_index=False)["In Need"].sum()
national_allocations = allocation_df.groupby("Country Code", as_index=False)["Total Allocation"].sum()

#Merge the two dataframes by country
final_total_df = national_needs.merge(
    national_allocations,
    left_on="Country ISO3",
    right_on="Country Code",
    how="inner"
)

#Divide Total Allocation by In Need
final_total_df["Total"] = final_total_df["Total Allocation"] / final_total_df["In Need"]

final_total_df
```
